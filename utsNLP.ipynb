```python
import re
import string
import nltk

from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')
```

```python
text = "KPK Sebut Nama Pengusaha Rokok Haji Her Ada di Temuan Dokumen Kasus Pejabat Bea Cukai Sumber: https://nasional.kompas.com/read/2026/04/13/22272291/kpk-sebut-nama-pengusaha-rokok-haji-her-ada-di-temuan-dokumen-kasus-pejabat?source=terpopuler..Membership: https://kmp.im/plus6 Download aplikasi: https://kmp.im/app6"
print("Teks asli:")
print(text)
```

```python
# 1. TEXT CLEANING
def text_cleaning(text):
    # hapus URL
    text = re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE)

    # hapus angka
    text = re.sub(r'\d+', '', text)

    # hapus tanda baca
    text = text.translate(str.maketrans('', '', string.punctuation))

    # hapus emoji / karakter non-ascii
    text = re.sub(r'[^\x00-\x7F]+', ' ', text)

    # hapus spasi berlebih
    text = re.sub(r'\s+', ' ', text).strip()

    return text

clean_text = text_cleaning(text)
print("1. Hasil Text Cleaning:")
print(clean_text)
```

```python
# 2. TOKENIZATION
tokens = word_tokenize(clean_text)

print("2. Hasil Tokenization:")
print(tokens)
```

```python
# 3. NORMALIZATION
normalization_dict = {
    "Gue": "saya",
    "gue": "saya",
    "gw": "saya",
    "nggak": "tidak",
    "gak": "tidak",
    "ga": "tidak",
    "bgt": "banget",
    "aja": "saja",
    "yg": "yang",
    "tdk": "tidak",
    "pelayanannya": "pelayanan",
    "harganya": "harga"
}

def normalization(tokens):
    return [normalization_dict[token] if token in normalization_dict else token for token in tokens]

normalized_tokens = normalization(tokens)
print("3. Hasil Normalization:")
print(normalized_tokens)
```

```python
# 4. STOPWORD REMOVAL
stop_words = set(stopwords.words('indonesian'))

def stopword_removal(tokens):
    return [word for word in tokens if word not in stop_words]

tokens_no_stopword = stopword_removal(normalized_tokens)
print("4. Hasil Stopword Removal:")
print(tokens_no_stopword)
```

```python
# 5. CASE FOLDING
def case_folding(tokens):
    return [token.lower() for token in tokens]

tokens_casefold = case_folding(tokens_no_stopword)
print("5. Hasil Case Folding:")
print(tokens_casefold)
```

```python
# 6. STEMMING
factory = StemmerFactory()
stemmer = factory.create_stemmer()

def stemming(tokens):
    return [stemmer.stem(token) for token in tokens]

stemmed_tokens = stemming(tokens_casefold)
print("6. Hasil Stemming:")
print(stemmed_tokens)
```

```python
# 7. LEMMATIZATION
# Lemmatization sederhana berbasis dictionary
lemma_dict = {
    "makanan": "makan",
    "pelayanan": "layan",
    "harga": "harga",
    "mahal": "mahal",
    "jelek": "jelek",
    "banget": "banget",
    "tidak": "tidak",
    "suka": "suka"
}

def lemmatization(tokens):
    return [lemma_dict[token] if token in lemma_dict else token for token in tokens]

lemmatized_tokens = lemmatization(stemmed_tokens)
print("7. Hasil Lemmatization:")
print(lemmatized_tokens)
```
